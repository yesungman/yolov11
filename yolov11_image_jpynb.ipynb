```python
# ▶️ 필수 라이브러리 설치
!pip install matplotlib pillow --quiet

# ▶️ 라이브러리 import
from PIL import Image
import matplotlib.pyplot as plt

# ▶️ 이미지 파일 경로
img1_path = "/mnt/data/화면 캡처 2025-05-07 153201.png"  # YOLOv8 그래프
img2_path = "/mnt/data/화면 캡처 2025-05-07 153655.png"  # YOLOv11 성능 표 및 그래프

# ▶️ YOLOv8 vs YOLOv11 성능 설명 출력
comparison_text = """
✅ YOLOv8 vs YOLOv11 성능 비교

1. 정확도(mAP):
   - YOLOv8: 최대 약 53~54 (COCO 기준)
   - YOLOv11: 최대 54.7 (YOLO11x 기준)
   → YOLOv11이 전체적으로 더 높은 정확도를 보여줍니다.

2. 추론 속도:
   - YOLOv8: A100 GPU 기준 약 1.0~3.0ms 수준
   - YOLOv11: T4 기준에서도 1.5~11ms 사이로 빠른 성능
   → YOLOv11은 다양한 환경에서 속도도 우수합니다.

3. 파라미터 수 및 구조:
   - YOLOv11은 더 작은 모델(n, s)부터 큰 모델(x)까지 파라미터 효율이 좋고, mAP 상승 폭도 큽니다.

아래는 두 모델의 성능 그래프입니다.
"""

print(comparison_text)

# ▶️ 이미지 출력 함수 정의
def show_image(path, title):
    img = Image.open(path)
    plt.figure(figsize=(12, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(title, fontsize=14)
    plt.show()

# ▶️ 이미지 출력
show_image(img1_path, "📈 YOLOv8 성능 그래프")
show_image(img2_path, "📈 YOLOv11 성능 그래프 및 표")
```
