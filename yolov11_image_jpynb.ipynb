```python
# ‚ñ∂Ô∏è ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install matplotlib pillow --quiet

# ‚ñ∂Ô∏è ÎùºÏù¥Î∏åÎü¨Î¶¨ import
from PIL import Image
import matplotlib.pyplot as plt

# ‚ñ∂Ô∏è Ïù¥ÎØ∏ÏßÄ ÌååÏùº Í≤ΩÎ°ú
img1_path = "/mnt/data/ÌôîÎ©¥ Ï∫°Ï≤ò 2025-05-07 153201.png"  # YOLOv8 Í∑∏ÎûòÌîÑ
img2_path = "/mnt/data/ÌôîÎ©¥ Ï∫°Ï≤ò 2025-05-07 153655.png"  # YOLOv11 ÏÑ±Îä• Ìëú Î∞è Í∑∏ÎûòÌîÑ

# ‚ñ∂Ô∏è YOLOv8 vs YOLOv11 ÏÑ±Îä• ÏÑ§Î™Ö Ï∂úÎ†•
comparison_text = """
‚úÖ YOLOv8 vs YOLOv11 ÏÑ±Îä• ÎπÑÍµê

1. Ï†ïÌôïÎèÑ(mAP):
   - YOLOv8: ÏµúÎåÄ ÏïΩ 53~54 (COCO Í∏∞Ï§Ä)
   - YOLOv11: ÏµúÎåÄ 54.7 (YOLO11x Í∏∞Ï§Ä)
   ‚Üí YOLOv11Ïù¥ Ï†ÑÏ≤¥Ï†ÅÏúºÎ°ú Îçî ÎÜíÏùÄ Ï†ïÌôïÎèÑÎ•º Î≥¥Ïó¨Ï§çÎãàÎã§.

2. Ï∂îÎ°† ÏÜçÎèÑ:
   - YOLOv8: A100 GPU Í∏∞Ï§Ä ÏïΩ 1.0~3.0ms ÏàòÏ§Ä
   - YOLOv11: T4 Í∏∞Ï§ÄÏóêÏÑúÎèÑ 1.5~11ms ÏÇ¨Ïù¥Î°ú Îπ†Î•∏ ÏÑ±Îä•
   ‚Üí YOLOv11ÏùÄ Îã§ÏñëÌïú ÌôòÍ≤ΩÏóêÏÑú ÏÜçÎèÑÎèÑ Ïö∞ÏàòÌï©ÎãàÎã§.

3. ÌååÎùºÎØ∏ÌÑ∞ Ïàò Î∞è Íµ¨Ï°∞:
   - YOLOv11ÏùÄ Îçî ÏûëÏùÄ Î™®Îç∏(n, s)Î∂ÄÌÑ∞ ÌÅ∞ Î™®Îç∏(x)ÍπåÏßÄ ÌååÎùºÎØ∏ÌÑ∞ Ìö®Ïú®Ïù¥ Ï¢ãÍ≥†, mAP ÏÉÅÏäπ Ìè≠ÎèÑ ÌÅΩÎãàÎã§.

ÏïÑÎûòÎäî Îëê Î™®Îç∏Ïùò ÏÑ±Îä• Í∑∏ÎûòÌîÑÏûÖÎãàÎã§.
"""

print(comparison_text)

# ‚ñ∂Ô∏è Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†• Ìï®Ïàò Ï†ïÏùò
def show_image(path, title):
    img = Image.open(path)
    plt.figure(figsize=(12, 6))
    plt.imshow(img)
    plt.axis('off')
    plt.title(title, fontsize=14)
    plt.show()

# ‚ñ∂Ô∏è Ïù¥ÎØ∏ÏßÄ Ï∂úÎ†•
show_image(img1_path, "üìà YOLOv8 ÏÑ±Îä• Í∑∏ÎûòÌîÑ")
show_image(img2_path, "üìà YOLOv11 ÏÑ±Îä• Í∑∏ÎûòÌîÑ Î∞è Ìëú")
```
